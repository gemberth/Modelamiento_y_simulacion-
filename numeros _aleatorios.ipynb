`**"Ninguna cantidad de pruebas puede probar que un GNA dado sea perfect

Prueba de frecuencia en bloques
El enfoque de esta prueba es la proporción de ** 1 ** dentro de bloques de ** m ** bits. El proósito
es determinar si una frecuencia de ** 1 ** dentro de un bloque es aproximadamente m/2, como se
esperaría de la aleatoreidad.
Parámetros
- m longitud de cda bloque
- n longitud de la lista de datos
- vector lista con la secuencia de bits 1 y 0
