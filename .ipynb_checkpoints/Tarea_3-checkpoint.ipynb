```python
# Generador de números aleatorios congruencia lineal
n, m, a, x0, c = 20, 1000, 101, 4, 457
x = [1] * n
r = [0.1] * n
# Se imprimen las variables utilizadas
print ("Método congruencial lineal")
print("----------------------------------")
print ("n = Cantidad de números generados\n")
print ("n : ", n)
print ("m : ", m)
print ("a : ", a)
print ("c : ", c)
print ("Xo : ", x0 )
for i in range(0, n):
    
    x[i] = ((a*x0)+c) % m
    x0 = x[i]
    r[i] = x0 / m

# llenamos nuestro DataFrame
d = {'Xn': x, 'ri': r }
df1 = pd.DataFrame(data=d)
df1
```

## Congruenciales multiplicativo

```python
# Generador de números aleatorios utilizando
# el método congruencial multiplicativo
n, m, a, x0 = 20, 30307, 172, 172
x = [1] * n
r = [0.1] * n
print (" Generador Congruencial multiplicativo")
print ("--------------------------------------")
for i in range(0, n):    
    x[i] = (a*x0) % m
    x0 = x[i]
    r[i] = x0 / m

# Creamos el dataframe y se imprime
d = {'Xn': x, 'ri': r }
df2 = pd.DataFrame(data=d)
df2
```

# 4.2 Test de números aleatorios

```python
# Test de números aleatorios con
# prueba de frecuencias en bloques
import math
from scipy.stats import chi2
# Extraemos los datos
rS = df2["ri"]
lista1=np.array(rS)
vector = []
# Llenamos de 0 y 1
for i in range (0,len(lista1)-1):
    
    if lista1[i] < lista1[i+1]:
               
        vector.append(0)
    else:
        
        vector.append(1)
# Vector
m = 10 # Número de bloques
n = len(vector) # longitud de la lista
N = int(n/m) # Número de elementos de cada bloque
blocks = []
i = 0
# Se aplica el test de frecuencias
while(len(blocks) != N):
    
    blocks.append(vector[i:(i+m)])
    i = i + m
sumas = []
for i in range(len(blocks)):    
    sumas.append(0)    
    for j in range(len(blocks[i])):        
        sumas[i] = sumas[i] + blocks[i][j]/float(m)
x_2 = 0
for i in range(N):
    x_2 = x_2 + math.pow((sumas[i] - .5),2)
x_2 = x_2 * 4*m
df = m - 1
p_value = chi2.cdf(x_2, df)
if p_value < 0.01:        
    print("Test de frecuencia en bloques:\nNo pasa\n")
else:
    print("Test de frecuencia en bloques:\nAprobado\n")
# Obtenemos los resultados
print("Valor de p_value es:", round(p_value, 6))
print("Valor de chi cuadrado es:", round(x_2, 6),"\n")
print("La proporción del bloque de sumas es:")
titulo = []
n=20
for i in range(n):    
    titulo.append(float(i))
# Se imprimen los resultados
print(titulo,"\n")
print("Sumas:")
print(round(sumas[0], 3))
```

# Distribución de Frecuencia Poisson

La distribución de Poisson se especifica por un parámetro: lambda (λ). Este parámetro es igual a
la media y la varianza. Cuando lambda aumente a valores lo suficientemente grandes, la
distribución normal (λ, λ) podría utilizarse para aproximar la distribución de Poisson. Utilice la
distribución de Poisson para describir el número de veces que un evento ocurre en un espacio


finito de observación. Por ejemplo, una distribución de Poisson puede describir el número de
defectos en el sistema mecánico de un avión o el número de llamadas a un centro de llamadas
en una hora. La distribución de Poisson se utiliza con frecuencia en el control de calidad, los
estudios de fiabilidad/supervivencia y los seguros. Una variable sigue una distribución de
Poisson si se cumplen las siguientes condiciones: Los datos son conteos de eventos (enteros no
negativos, sin límite superior). Todos los eventos son independientes. La tasa promedio no
cambia durante el período de interés.


```python
import numpy as np
import scipy.stats as stats
from scipy.stats import poisson
import matplotlib.pyplot as plt
landa = 5
poisson = stats.poisson(landa)
x = np.arange(poisson.ppf(0.01), poisson.ppf(0.99))
fmp = poisson.pmf(x)
x
```

```python
fmp
```

```python
#graficamos con los valores generados
plt.plot(x,fmp,'--')
plt.vlines(x,0,fmp,colors='b',lw=5,alpha=0.5)
plt.title('Distribución de Poisson')
plt.ylabel('Probabilidad')
plt.xlabel('Valores')
plt.show()
```
