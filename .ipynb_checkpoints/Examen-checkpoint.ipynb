```python
#importamos las librerias a utilizar 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
# para esto imprortamos un data set de formato json 
datos= pd.read_excel('examen.xlsx')
datos

```

# a) Genere una tabla con el histograma de los datos con el intervalo, frecuencia observada, frecuencia relativa, frecuencia acumulada


```python
lis = datos["Tiempo"].unique()
lis
dfclases=pd.DataFrame(lis,columns=["tiempo"])
datafi=pd.crosstab(index=datos["Tiempo"], columns = "Fi")
li= datafi.values
dfclases["Fi"] = li
total = dfclases.sum(axis=0)
datahi = dfclases["Fi"]/total["Fi"]
datahi.values
dfclases["hi"] = datahi
dfclases
```

```python
x=datos["Tiempo"]
plt.figure(figsize=(10,5))
plt.hist(x,bins=8,color='gray')
plt.axvline(x.mean(),color='red',label='Media')
plt.axvline(x.median(),color='yellow',label='Mediana')
plt.axvline(x.mode()[0],color='green',label='Moda')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia Relativa')
plt.legend()
plt.show()
```

# Dibujar la frecuencia acumulada

```python
# Calculamos la frecuencia acumulada
datafi=pd.crosstab(index=datos['Tiempo'], columns = "fi")
# Creamos una lista con los valores de las frecuencias
li = datafi.values
# Agregamos una columna al dataframe
dfclases["fi"] = li
# Observamos dfclase
dfclases
```

```python
x=datos["Tiempo"]
plt.figure(figsize=(10,5))
plt.hist(x,bins=None,color='gray')
plt.axvline(x.mean(),color='red',label='Media')
plt.axvline(x.median(),color='yellow',label='Mediana')
plt.axvline(x.mode()[0],color='green',label='Moda')
plt.xlabel('Tiempo')
plt.ylabel('Frecuencia acumulada')
plt.legend()
plt.show()
```

# b) Grafique la funciÃ³n fdp((frecuencia relativa) y FDA (frecuencia acumulada)

```python
FA = dfclases["fi"].values
a=[]
b=0
for c in FA:   
    
    b = c + b
    a.append(b)
dfclases["FA"] = a
HI = dfclases["hi"].values
a=[]
b=0
for c in HI:
    b = c + b
    a.append(b)
dfclases["HI"] = a
dfclases

```

```python
x=dfclases["tiempo"]
t1=dfclases["hi"]
t2=dfclases["HI"]
plt.figure(figsize=(9,4))
plt.plot(x,t1,x,t2,marker='o')
plt.xlabel('segundos')
plt.ylabel('FRECUENCIA')
plt.legend(('frec absoluta','frec acumulada'),prop={'size':10},loc='upper right')
```

# d)   Genere 10 nÃºmeros aleatorios con los siguientes parÃ¡metros Xo=349, a=347,c=47,M=101

```python
n = 10
x0 = 349
a = 347
c = 47
m = 101
x = [1] * n
r = [0.1] * n
for i in range(0, n):    
    x[i] = ((a*x0)+c) % m
    x0 = x[i]
    r[i] = x0 / m
d = {'Valores': x, 'Tiempo': r }
df1 = pd.DataFrame(data=d)
# Se imprime el resultado ordenado
precio = df1.sort_values('Tiempo')
precio.style.hide_index()
```
