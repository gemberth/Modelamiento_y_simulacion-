```python
# Generamos los 62 # aleatorios por cualquiera de los mÃ©todos estudiados y luego pode
#mos realizar la simulaciÃ³n
# Generador de nÃºmeros aleatorios Congruencia lineal
# Borland C/C++ xi+1=22695477xi + 1 mod 2^32
n, m, a, x0, c = 600, 2**32, 22695477, 4, 1
x = [1] * n
r = [0.1] * n
for i in range(0, n):
    x[i] = ((a*x0)+c) % m
    x0 = x[i]
    r[i] = x0 / m
# llenamos nuestro DataFrame
d = {'ri': r }
dfMCL = pd.DataFrame(data=d)
dfMCL
```
